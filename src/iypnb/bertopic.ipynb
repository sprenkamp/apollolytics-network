```python
from bertopic import BERTopic
from sklearn.datasets import fetch_20newsgroups
import pandas as pd
from deep_translator import GoogleTranslator

def translate(x):
    try:
        tranlsation = GoogleTranslator(source='auto', target='en').translate(x)
    except:
        tranlsation = ""
    return tranlsation


df = pd.read_csv("../../data/telegram/telegram.csv")
df= df.sample(10_000)
df['englishMessageText'] = df.messageText.apply(lambda x: translate(x))

docs = df.englishMessageText.fillna("").values.tolist()

topic_model = BERTopic(language="multilingual")
topics, probs = topic_model.fit_transform(docs)
topic_model.get_topic_info()
```
